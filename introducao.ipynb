####2 - Quais dados vão me ajudar a resolver o problema? (80/20)

- Aqui, você deve pensar nos '80/20', ou seja, quais são os fatores mais relevantes que influenciam os dados/indicadores
- Esse 80/20 é famoso 'princípio de pareto'
- O italiano Vilfredo Pareto percebeu que 80% das terras na Itália perteciam a 20% da população. Assim como 80% dos frutos de jardim dele eram gerados por 20% das plantas
- Mais tarde, esse conceito se popularizou no contexto do dia das empresas
- O princípio de Pareto ou a Regra 80/20 diz que 80% dos resultados vêm de 20% das causas principais. Claro que os números exatos podem variar, mas a ideia central é que existe uma quantidade pequena de fatores que pode gerar a maior parte dos resultados.

####Exemplos:

- Em muitos países, 80% da riqueza pertecem a 20% da população
- 80% das vendas de uma empresa geralmente vêm de 20% clientes
- 80% do trabalho em um projeto muitas vezes é realizado por 20% da equipe
- 80% das visitas de um site podem ser geradas por 20% das campanhas de marketing
- 80% dos custos de manutenção vêm de 20% dos equipamentos

Então pensando no nosso cenário de Análise de Dados:

- 80% dos insights úteis podem vir de 20% dos dados analisados
- 80% dos problemas identificados podem ser atribuídos a 20% das causas.

Então aplicando no nosso caso: vamos pensar no nosso 80/20, quais são os fatores mais relevantes que influenciam os dados/indicadores pra aumentar a lucratividade da Coca-Cola

Exemplos:
- Receita mensal
- Receita por produto
- Custo mensal
- Abertura dos custos

####5 - Questionar e entender SEMPRE

- Entrando na raiz do problema, dá pra perceber que é a matéria prima da bebida. Mais específicamente, açúcar e plástico
- Conseguimos trocar de fornecedor? Reduzir a quantidade dos insumos sem afetar a qualidade?
- Resultado: não conseguimos trocar de fornecedor de açúcar, que é um no Brasil, nem reduzir a quantidade sem perder qualidade
- Resultado: 'Coca-Cola lança primeira garrafa de água mineral feita 100% com resina reciclada'. Deu pra cortar pela metade a quantidade de plástico de uma das principais linhas de custo da fábrica (e ainda usa como branding)

    - Ex: 
    - Foi notado que em países do hemisfério norte que as pessoas tendem a gastar mais em compras durante a época de frio
